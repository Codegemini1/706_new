# BREAK, CONTINUE, PASS

# BREAK

```python
for i in range(1,6):
    if i == 3:
        break
    print(i)
```

```python
for i in 'python':
    if i == 'o':
        break
    print(i)
```

# Continue SKIP

```python
for i in range(1,6):
    if i == 3:
        continue
    print(i)
```

```python
for i in 'python':
    if i == 'o':
        continue
    print(i)
```

```python
for i in 'hello':
    if i == 'l':
        continue
    print(i)
```
