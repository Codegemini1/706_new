```python
def square(n):
    l = []
    for i in range(1,n+1):
        l.append(i * i)
    return l
```

```python
square(5)
```

```python
def square_gen(n):
    for i in range(1,n+1):
        yield i * i
```

```python
for square in square_gen(8):
    print(square)
```
