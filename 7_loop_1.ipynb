```python
l = [1,1.5,[10,2.5,[20,3.5],20],19.2]
```

```python
for i in l:
    if type(i) == float:
        print(i)
    elif type(i) == list:
        for l1 in i:
            if type(l1) == float:
                print(l1)
            elif type(l1) == list:
                for l2 in l1:
                    if type(l2) == float:
                        print(l2)
```
