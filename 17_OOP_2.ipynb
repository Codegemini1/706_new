```python
class banking():
    def __init__(self,balance):
        self.balance = balance

class hdfc(banking):
    def mutual_funds(self):
        result = self.balance/10
        return result

class axis(hdfc):
    def buy_stocks(self,number_shares):
        if number_shares * 5 <= self.balance:
            return f'Total number of shares is now {number_shares} and total investment is {number_shares * 5}'
        else:
            return 'Insufficient funds'
```

```python
c1 = axis(20)
```

```python
c1.mutual_funds()
```

```python
c1.buy_stocks(10)
```

```python
c1.buy_stocks(2)
```

```python
moti = dog()
```

```python
moti.sound()
```

```python
c1.sound()
```

# Abstraction Method

```python
from abc import ABC , abstractmethod
```

```python
class car(ABC):
    @abstractmethod
    def airbag(self):
        pass
```

```python
class nexon(car):
    def airbag(self):
        return 'Air bags are installed in tata nexon'
    # optional
    def sunroof(self):
        return 'Sunroof is installed in tata nexon'
```

```python
c10  = nexon()
```

```python
c10.airbag()
```

```python
c10.sunroof()
```

```python
class fortuner(car):
    def autopark(self):
        return 'Autopark feature added'
```

```python
f = fortuner()
```

```python
class phone(ABC):
    @abstractmethod
    def call(self):
        pass
```

```python
class iphone(phone):
    def call(self):
        return 'Calling feature is added in iphone'
    def camera(self):
        return 'camera is added in iphone'
```

```python
iphone13 = iphone()
```

```python
iphone13.call()
```

```python
iphone13.camera()
```
