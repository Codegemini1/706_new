```python
# function for continous data
def test4(*args):
    return args
```

```python
test4(1,3,'Hello', True, 'By')
```

```python
def test5(*args):
    k = []
    for i in args:
        if type(i) == list:
            for data in i:
                k.append(data)
    return k
```

```python
test5(1,2,3,[10,20,'Hello'],True,'Python',['A','B','C',100])
```
