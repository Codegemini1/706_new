```python
def square(a):
    return a*a
```

```python
def n_square(*args):
    for i in args:
        print(i * i)
```

```python
n_square(3,2,4,14,5,24,6)
```

```python
def test(n):
    for i in range(1,n):
        print(i*i)
```

```python
test(10)
```

```python
def test2(n):
    for i in range(1,n+1):
        yield i * i
```

```python
for i in test2(10):
    print(i)
```
