```python
class house_price():
    def __init__(self,size,house_type,floors,lift):
        self.size = size
        self.house_type = house_type
        self.floors = floors
        self.lift = lift
    def price(self):
        result = self.size * 4000
        print(result)
        
```

```python
mannat = house_price(4000,'bunglow',3,'yes')
```

```python
mannat.price()
```

```python
class new_car():
    def __init__(self,color,engine_type,doors): # constructor
        self.color = color
        self.engine_type = engine_type
        self.doors = doors
    def price(self,location):
        if location == 'pune':
            print(f'Your car with color {self.color} is availabel in pune')
        else:
            print(f'Your car is not avilable in your location {location}')
```

```python
punch = new_car('White', 'Petrol', 4)
```

```python
punch.price('pune')
```

```python
punch.price('mumbai')
```
