```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.relplot(data = df, x= 'tip', y = 'total_bill', hue = 'size')
```

```python
sns.relplot(data = df, x= 'tip', y = 'total_bill', hue = 'smoker', style='time')
```

```python
sns.relplot(data = df, x= 'tip', y = 'total_bill', hue = 'sex', style='size')
```

```python
df['smoker'].value_counts()
```

```python
sns.countplot(data = df, x='smoker')
```

```python
sns.countplot(data = df, x = 'size')
```

```python
sns.lineplot(data = df, x = 'size', y = 'total_bill')
plt.title('Bill vs Tip')
plt.show()
```

```python
sns.histplot(df['total_bill'], bins=15)
plt.title('Bill distribution')
plt.show()
```

```python
sns.boxplot(data = df, x= 'total_bill')
```

```python
# Corr 
df[['total_bill','tip','size']].corr()
```
